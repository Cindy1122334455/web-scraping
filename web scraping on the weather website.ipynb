Crawling the weather data of New York City in the coming week based on the website “NATIONAL WEATHER”, https://forecast.weather.gov/MapClick.phplat=40.7146&lon=-74.0071#.XBBh5xNKg_U

```python
#using urllib module to return the content of the weather website
import urllib
weather_url='https://forecast.weather.gov/MapClick.php?lat=40.7146&lon=-74.0071#.XBBh5xNKg_U'
web_page=urllib.urlopen(weather_url).read()
print(web_page)
```

```python
#using BeautifulSoup to grab weather information
from bs4 import BeautifulSoup
soup=BeautifulSoup(web_page,'html.parser')
print(soup.find(id='seven-day-forecast-body').get_text())
```

```python
#Output a beautiful website via prettify()
from bs4 import BeautifulSoup
soup=BeautifulSoup(web_page,'html.parser')
print(soup.find(id='seven-day-forecast-container').prettify())
```

```python
#Extract data and keep data completely and orderly
soup_forecast=soup.find(id='seven-day-forecast-container')
date_list=soup_forecast.find_all(class_='period-name')
desc_list=soup_forecast.find_all(class_='short-desc')
temp_list=soup_forecast.find_all(class_='temp')
for i in range(9):
    date=date_list[i].get_text()
    desc=desc_list[i].get_text()
    temp=temp_list[i].get_text()
    print(u"{}, {}, {}".format(date, desc, temp))
```
